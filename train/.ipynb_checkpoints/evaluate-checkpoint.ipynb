# Plot Functions

```python
def plot_label(r1, r2, pred_label, img, figsize=(15,10)):
    r1 = np.ma.masked_where(r1 <= 0.01, r1)
    r2 = np.ma.masked_where(r2 <= 0.01, r2)
    pred_label = np.ma.masked_where(pred_label <= 0.05, pred_label)

    fig, ax = plt.subplots(1,3, figsize=figsize)
    ax[0].imshow(img, cmap=plt.get_cmap('binary'))
    ax[0].imshow(r1, cmap=plt.get_cmap('Reds'), alpha=0.8, vmin=0, vmax=1)
    ax[0].set_title("Rater 1")
    ax[1].imshow(img, cmap=plt.get_cmap('binary'))
    ax[1].imshow(r2, cmap=plt.get_cmap('Reds'), alpha=0.8, vmin=0, vmax=1)
    ax[1].set_title("Rater 2")
    ax[2].imshow(img, cmap=plt.get_cmap('binary'))
    ax[2].imshow(pred_label, cmap=plt.get_cmap('Reds'), alpha=0.8, vmin=0, vmax=1)
    ax[2].set_title("Prediction")
    
def plot_label1(rater, pred_label, img, figsize=(15,10)):
    rater = np.ma.masked_where(rater <= 0.01, rater)
    pred_label = np.ma.masked_where(pred_label <= 0.05, pred_label)

    fig, ax = plt.subplots(1,2, figsize=figsize)
    ax[0].imshow(img, cmap=plt.get_cmap('binary'))
    ax[0].imshow(rater, cmap=plt.get_cmap('Reds'), alpha=0.6, vmin=0, vmax=1)
    ax[0].set_xticks([], [])
    ax[0].set_yticks([], [])
    ax[1].imshow(img, cmap=plt.get_cmap('binary'))
    ax[1].imshow(pred_label, cmap=plt.get_cmap('Reds'), alpha=0.6, vmin=0, vmax=1)
    ax[1].set_xticks([], [])
    ax[1].set_yticks([], [])
    plt.tight_layout()
    
def plot_label2(rater, pred_label, img_t1, img_t2, img_flair, figsize=(15,10)):
    #rater = np.ma.masked_where(rater, rater)
    #pred_label = np.ma.masked_where(pred_label, pred_label)

    fig, ax = plt.subplots(1,5, figsize=figsize)
    for a in ax:
        a.set_xticks([], [])
        a.set_yticks([], [])
    
    ax[0].imshow(img_t1, cmap=plt.get_cmap('binary'))
    ax[1].imshow(img_t2, cmap=plt.get_cmap('binary'))
    ax[2].imshow(img_flair, cmap=plt.get_cmap('binary'))

    #ax[3].imshow(img_t1, cmap=plt.get_cmap('binary'))
    ax[3].imshow(rater, cmap=plt.get_cmap('Reds'), alpha=0.8, vmin=0, vmax=1)

    #ax[4].imshow(img_t1, cmap=plt.get_cmap('binary'))
    ax[4].imshow(pred_label, cmap=plt.get_cmap('Reds'), alpha=0.8, vmin=0, vmax=1)

    plt.tight_layout()
```

## Utils

```python
def loss_and_score(outputs, r1, r2):
    loss_r1 = dice_loss(outputs, r1)
    loss_r2 = dice_loss(outputs, r2)
    loss = (loss_r1 + loss_r2 ) * 0.5
    # metric
    score_r1, _ = dice_metric(post_inference(outputs), r1)
    score_r2, _ = dice_metric(post_inference(outputs), r2)
    score = (score_r1 + score_r2 ) * 0.5
    
    return loss, score
```

```python
post_inference = Compose([Activations(softmax=True), AsDiscrete(threshold_values=True)])
dice_loss = DiceLoss(softmax=True, include_background=False)
gen_dice_loss = GeneralizedDiceLoss(softmax=True)
dice_metric = DiceMetric(include_background=False)
softmax = Activations(softmax=True)
discretize = AsDiscrete(threshold_values=True)
metric = DiceMetric(include_background=False, reduction="mean")
sigmoid = Activations(sigmoid=True)
bce_loss = BCELoss()
device = 'cpu'
```
