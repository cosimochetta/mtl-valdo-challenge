# Utils

```python
def get_file_list_task3(root_dir):
    file_list = glob(root_dir + "/**/*", recursive=True)
    file_list = [file for file in file_list if '.nii.gz' in file]

    fl_T1 = sorted([file for file in file_list if '_T1' in file])
    fl_T2 = sorted([file for file in file_list if '_T2' in file])
    fl_FLAIR = sorted([file for file in file_list if '_FLAIR' in file])
    fl_R1 = sorted([file for file in file_list if 'Rater1' in file or 'Rater3' in file])
    fl_R2 = sorted([file for file in file_list if 'Rater1' in file or 'Rater4' in file])
    
    assert len(fl_T1) == len(fl_T2) == len(fl_FLAIR) == len(fl_R1) == len(fl_R2)
    return fl_T1, fl_T2, fl_FLAIR, fl_R1, fl_R2


def load_nifty_from_file(path, loader='nib'):
    if loader == 'nib':
        nifti_orig = nib.load(path)
        return nifti_orig.get_data()  # transform the images into np.ndarrays
    raise NameError('No loader found')
    
def create_task3_mask(data):
    '''
    Consider median of data as background because yes
    '''
    bg_value = np.median(data)
    mask = (data != bg_value).astype(int)
    return mask

def cut_outliers_task3(data, cut_value):
    '''
    Consider median of data as background because yes
    '''

    bg_value = np.median(data)
    mask = np.logical_or(data <= bg_value, data > cut_value)
    ma = np.ma.array(data, mask=mask)
    mean = ma.mean()
    
    data[data > cut_value] = mean
    return data
    
def rescale_data(data, mask, min_value=None, max_value=None):
    '''
    rescale data between a fixed range, ignore mask data
    '''
    
    # Invert mask for masked array
    mask = mask == 0
    
    ma = np.ma.array(data, mask=mask)
    
    if not min_value:
        min_value = ma.min()
    if not max_value:
        max_value = ma.max()
    
    ma = (ma - min_value) / (max_value - min_value)
    data = ma.filled(0)
    return data
    
def rescale_labels(label):
    if label.min() == 0 and label.max() <= 1:
        return label
    
    mask = label == 0
    ma = np.ma.array(label, mask=mask)

    min_value = ma.min()
    max_value = ma.max()
    
    ma = (ma - min_value) / (max_value - min_value)
    label = ma.filled(0)
    
    return label 
```

```python
class ValdoDatasetTask3Patches(Dataset):

    def __init__(self, root_dir, split="Train", patch_size=96):
        # Define attributes
        self.root_dir = root_dir
        self.data_list = []
        self.split = split
        
        self.to_tensor = Compose([
                AddChannel(),
                ToTensor()
        ])
        
        # Load data index
        data_dir = os.path.join(root_dir, str(patch_size))
        t1_paths, t2_paths, flair_paths, r1_paths, r2_paths = get_file_list_task3_patches(data_dir)
        self.t1_data = np.load(t1_paths[0])
        self.t2_data = np.load(t2_paths[0])
        self.flair_data = np.load(flair_paths[0])
        self.r1_data = np.load(r1_paths[0])
        self.r2_data = np.load(r2_paths[0])
        
        print("Loading data...")
        for i in range(1, len(t1_paths)):
            print(f"{i} / {len(t1_paths)}", end="\r")
            t1 = np.load(t1_paths[i])
            t2 = np.load(t2_paths[i])
            flair = np.load(flair_paths[i])
            r1 = np.load(r1_paths[i])
            r2 = np.load(r2_paths[i])

            self.t1_data = np.concatenate((self.t1_data, t1), axis=0)
            self.t2_data = np.concatenate((self.t2_data, t2), axis=0)
            self.flair_data = np.concatenate((self.flair_data, flair), axis=0)
            self.r1_data = np.concatenate((self.r1_data, r1), axis=0)
            self.r2_data = np.concatenate((self.r2_data, r2), axis=0)
            
            
                
    def __len__(self):
        return len(self.t1_data)

    def __getitem__(self, idx):
        t1 = self.to_tensor(self.t1_data[idx])
        t2 = self.to_tensor(self.t2_data[idx])
        flair = self.to_tensor(self.flair_data[idx])
        r1 = self.to_tensor(self.r1_data[idx])
        r2 = self.to_tensor(self.r2_data[idx])
        img = torch.cat((t1, t2, flair), dim=0)
                
        label = (r1 + r2) / 2
        label = torch.vstack((1-label, label))
        return {'t1': t1, 'img': img, 'r1': r1, 'r2': r2, 'label': label}

class DatasetOnlyDataWithLacunes(Dataset):

    def __init__(self, root_dir, split="Train", patch_size=96):
        # Define attributes
        self.root_dir = root_dir
        self.data_list = []
        self.split = split
        
        self.to_tensor = Compose([
                AddChannel(),
                ToTensor()
        ])
        
        # Load data index
        data_dir = os.path.join(root_dir, str(patch_size))
        t1_paths, t2_paths, flair_paths, r1_paths, r2_paths = get_file_list_task3_patches(data_dir)
        self.t1_data = np.load(t1_paths[0])
        self.t2_data = np.load(t2_paths[0])
        self.flair_data = np.load(flair_paths[0])
        self.r1_data = np.load(r1_paths[0])
        self.r2_data = np.load(r2_paths[0])
        
        print("Loading data...")
        for i in range(1, len(t1_paths)):
            print(f"{i} / {len(t1_paths)}", end="\r")
            t1 = np.load(t1_paths[i])
            t2 = np.load(t2_paths[i])
            flair = np.load(flair_paths[i])
            r1 = np.load(r1_paths[i])
            r2 = np.load(r2_paths[i])
            
            slices_to_remove = []
            for i, (r1s, r2s) in enumerate(zip(r1, r2)):
                if r1s.max() == 0 and r2s.max() == 0:
                     slices_to_remove.append(i)
            t1 = np.delete(t1, slices_to_remove, axis=0)
            t2 = np.delete(t2, slices_to_remove, axis=0)
            flair = np.delete(flair, slices_to_remove, axis=0)
            r1 = np.delete(r1, slices_to_remove, axis=0)
            r2 = np.delete(r2, slices_to_remove, axis=0)
            self.t1_data = np.concatenate((self.t1_data, t1), axis=0)
            self.t2_data = np.concatenate((self.t2_data, t2), axis=0)
            self.flair_data = np.concatenate((self.flair_data, flair), axis=0)
            self.r1_data = np.concatenate((self.r1_data, r1), axis=0)
            self.r2_data = np.concatenate((self.r2_data, r2), axis=0)
                
    def __len__(self):
        return len(self.t1_data)

    def __getitem__(self, idx):
        t1 = self.to_tensor(self.t1_data[idx])
        t2 = self.to_tensor(self.t2_data[idx])
        flair = self.to_tensor(self.flair_data[idx])
        r1 = self.to_tensor(self.r1_data[idx])
        r2 = self.to_tensor(self.r2_data[idx])
        img = torch.cat((t1, t2, flair), dim=0)
                
        label = (r1 + r2) / 2
        label = torch.vstack((1-label, label))
        return {'t1': t1, 'img': img, 'r1': r1, 'r2': r2, 'label': label}
```

```python
def get_file_list_task3_patches(root_dir):
    file_list = glob(root_dir + "/**/*", recursive=True)
    file_list = [file for file in file_list if '.npy' in file]

    fl_T1 = sorted([file for file in file_list if '_t1' in file])
    fl_T2 = sorted([file for file in file_list if '_t2' in file])
    fl_FLAIR = sorted([file for file in file_list if '_flair' in file])
    fl_R1 = sorted([file for file in file_list if 'r1' in file])
    fl_R2 = sorted([file for file in file_list if 'r2' in file])
    
    assert len(fl_T1) == len(fl_T2) == len(fl_FLAIR) == len(fl_R1) == len(fl_R2)
    return fl_T1, fl_T2, fl_FLAIR, fl_R1, fl_R2
```

```python
 # create a validation data loader
val_ds = DatasetOnlyDataWithLacunes(root_dir=val_path, 
                             patch_size=patch_size, 
                            split="val")
val_loader = DataLoader(
    val_ds,
    batch_size=64,
    num_workers=0,
    collate_fn=list_data_collate,
    pin_memory=torch.cuda.is_available(),
)
```

```python
### Check evaluation
dice_metric = DiceLoss(softmax=True, include_background=False)
n = len(val_loader)

i = 1
with torch.no_grad():
    score_sum = 0.0
    score_count = 0
    for val_data in val_loader:
        val_images, val_labels = val_data["t1"].to(device), val_data["label"].to(device)
        outputs = model(val_images)   
        loss = dice_metric(outputs, val_labels)
        score_count += 1
        score_sum += 1 - loss.item()
        print(f"Batch {i} / {n}: {1 - loss}")
        i+=1
    score = score_sum / score_count
    print(f"Score: {score}")
```
